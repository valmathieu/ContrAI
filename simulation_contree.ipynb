# 🃏 Étude Statistique à la Contrée : Le Cas de l'Égalité Pique/Cœur

L'objectif de ce notebook est de démontrer par la théorie, la visualisation et l'expérience la probabilité que les deux plus petites cartes d'une donne à 4 joueurs soient **un Pique et un Cœur de la même valeur**.

## 1. Approche Théorique

Le jeu compte $32$ cartes. Lors de la donne, on en tire $4$. 
Le nombre total de tirages possibles est donné par le coefficient binomial :
$$\binom{32}{4} = \frac{32 \times 31 \times 30 \times 29}{4 \times 3 \times 2 \times 1} = 35960$$

En filtrant les combinaisons pour ne garder que celles où les deux cartes les plus faibles sont un Pique et un Cœur de la même hauteur, on trouve exactement **1064 combinaisons gagnantes**.

La probabilité théorique est donc de $\frac{1064}{35960} \approx 2.9588\%$.

```python
import plotly.graph_objects as go

# Définition des nœuds (les étapes)
# Note : Les étiquettes incluent les probabilités théoriques calculées pour chaque branche
nodes = [
    "Tous les Tirages<br>(100%)",                        # Nœud 0 (Origine)
    "Condition A : Échouée<br>(La plus basse n'est pas ♥️)",# Nœud 1 (Échec A)
    "Condition A : Remplie<br>(La plus basse est ♥️)",    # Nœud 2 (Continuer)
    "Condition B : Échouée<br>(Suivante n'est pas ♠️)",    # Nœud 3 (Échec B)
    "<b>SUCCÈS 🏆</b><br>(Deux plus basses sont ♠️ & ♥️)"  # Nœud 4 (Succès Final)
]

# Définition des liens (le flux)
# Les 'values' représentent les proportions théoriques des chemins
# Ces proportions découlent de l'analyse combinatoire du problème.
# Note : Les échecs sont regroupés pour plus de clarté visuelle.

sources = [0, 0, 2, 2]  # Origines des flux (ID des nœuds)
targets = [1, 2, 3, 4]  # Destinations des flux (ID des nœuds)
values  = [76.8, 23.2, 20.2, 3.0] # Proportions (%) du flux entrant
colors  = ['#ecf0f1', '#3498db', '#ecf0f1', '#2ecc71'] # Couleurs (Gris/Bleu/Gris/Vert)

# Création de la figure Sankey
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 20,           # Espacement vertical entre les nœuds
      thickness = 30,     # Épaisseur des nœuds
      line = dict(color = "black", width = 0.5),
      label = nodes,      # Étiquettes définies ci-dessus
      color = ["#7f8c8d", "#e74c3c", "#3498db", "#e74c3c", "#2ecc71"] # Couleurs des nœuds
    ),
    link = dict(
      source = sources,   # Indices d'origine
      target = targets,   # Indices de destination
      value = values,     # Proportions
      color = colors,     # Couleurs des flux
      # Personnalisation du texte au survol de la souris
      hovertemplate = 'Flux de "%{source.label}" vers "%{target.label}" :<br><b>%{value}% des tirages</b><extra></extra>'
    ))])

# Finitions du layout
fig.update_layout(
    title_text="Cheminement logique vers le succès (Diagramme de Sankey interactif)",
    title_font_size=16,
    font_size=12,
    width=900, height=500,
    margin=dict(t=60, l=20, r=20, b=20),
    template="plotly_white"
)

fig.show()
```

## 2. Expérience Pratique : 4 Tirages pas à pas

Conformément à l'objectif, nous allons réaliser **exactement 4 tirages** aléatoires pour observer le comportement de la donne. L'ordre des couleurs de la plus forte à la plus faible est : Pique (4) > Cœur (3) > Carreau (2) > Trèfle (1).

```python
import random

# Préparation du paquet
valeurs = {7: '7', 8: '8', 9: '9', 10: '10', 11: 'Valet', 12: 'Dame', 13: 'Roi', 14: 'As'}
couleurs = {4: 'Pique ♠', 3: 'Coeur ♥', 2: 'Carreau ♦', 1: 'Trefle ♣'}

paquet = [(v, c) for v in valeurs.keys() for c in couleurs.keys()]

def verifier_succes(tirage):
    # Tri par valeur puis par couleur
    tirage_trie = sorted(tirage, key=lambda x: (x[0], x[1]))
    c1, c2 = tirage_trie[0], tirage_trie[1]
    
    # Si même valeur et que les couleurs sont Pique (4) et Coeur (3)
    if c1[0] == c2[0] and {c1[1], c2[1]} == {4, 3}:
        return True, tirage_trie
    return False, tirage_trie

print("=== LANCEMENT DES 4 TIRAGES ===")
succes_obtenus = 0

for i in range(1, 5):
    # On tire 4 cartes
    tirage_actuel = random.sample(paquet, 4)
    est_succes, main_triee = verifier_succes(tirage_actuel)
    
    # Formatage lisible
    main_texte = [f"{valeurs[c[0]]} de {couleurs[c[1]]}" for c in main_triee]
    
    print(f"\nTirage {i}:")
    print(f"Cartes (de la plus faible à la plus forte) : {', '.join(main_texte)}")
    if est_succes:
        print("✅ RÉSULTAT : SUCCÈS ! Les deux plus faibles sont un Pique et un Cœur de même valeur.")
        succes_obtenus += 1
    else:
        print("❌ RÉSULTAT : Échec.")

print(f"\nBilan sur 4 tirages : {succes_obtenus} succès.")
```

## 3. Visualisation de la convergence (Loi des Grands Nombres)

Faire 4 tirages est très instructif, mais avec une probabilité théorique de $\approx 3\%$, il est tout à fait normal de n'obtenir que des échecs (ou d'avoir beaucoup de chance et de surperformer). 

Pour observer la convergence vers les $2.96\%$, nous allons simuler $5000$ tirages et tracer un **graphique interactif**. Vous pouvez zoomer sur la courbe ou passer votre souris pour voir les valeurs exactes !

```python
import plotly.express as px
import pandas as pd

# Simulation sur 5000 tirages pour le graphique
N = 5000
historique_probabilites = []
succes_cumules = 0

for i in range(1, N + 1):
    tirage_actuel = random.sample(paquet, 4)
    if verifier_succes(tirage_actuel)[0]:
        succes_cumules += 1
    
    # On enregistre la probabilité calculée à l'instant i
    historique_probabilites.append((succes_cumules / i) * 100)

# Création d'un DataFrame (tableau) pour Plotly
df = pd.DataFrame({
    'Nombre de tirages': list(range(1, N + 1)),
    'Probabilité mesurée (%)': historique_probabilites
})

# Tracé du graphique interactif
fig = px.line(df, x='Nombre de tirages', y='Probabilité mesurée (%)', 
              title=f"Convergence de la probabilité sur {N} tirages")

# Ajout de la ligne cible théorique
fig.add_hline(y=2.9588, line_dash="dash", line_color="red", 
              annotation_text="Probabilité Théorique (2.96%)", 
              annotation_position="bottom right",
              annotation_font=dict(color="red", size=12))

# Finitions esthétiques
fig.update_layout(
    template="plotly_white",
    yaxis=dict(range=[0, 10]), # On fixe l'axe Y entre 0 et 10% pour y voir clair
    hovermode="x unified"
)

fig.show()
```
